```python
#####################################################################
# 3.   - / ,  ,   #
#####################################################################

#        (*****)

#            
#  .  (80%) (20%)   
#   (20%)      
#    . 

#  () () . 

X_data = df.loc[:,'sepal_width':'petal_width']   # ()  X 
y_data = df.loc[:,'target']                       # ()  y 

# by 
# X_data = df.iloc[:, [0,1,2,3]]   #  ()
# y_data = df.iloc[:, [4]]         #  ()
```

```python
# !!!   !!!

from sklearn.model_selection import train_test_split #     

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,
                                                    test_size = 0.2,
                                                    shuffle = True,
                                                    random_state = 20
                                                    )
```

```python
########################################################################
# 4.       - /  ,  , #
########################################################################

# KNN - K Nearest Neighbors (K)     
#            K  .  
#           K       
#              .  

#         k=3  Class B   2  . 
#        k=6  Class A   4  .
#        K  KNN      . 
#       K      K  . 
```

knn.png

```python
#   ( :    )

from sklearn.neighbors import KNeighborsClassifier   # KNN  
knn = KNeighborsClassifier(n_neighbors=20)            # knn  (k  )
knn.fit(X_train, y_train)  #       

#  ()

knn_pred = knn.predict(X_test)   #     y_test  
knn_pred

# 
from sklearn.metrics import accuracy_score
accuracy_score(y_test, knn_pred)

# n_neighbors=3     0.9333333333333333
# n_neighbors=6     0.9333333333333333
# n_neighbors=10    0.9333333333333333
# n_neighbors=20    0.9666666666666667   ->    1.0  'sepal_width':'petal_width' 
# n_neighbors=50    0.9333333333333333
```
